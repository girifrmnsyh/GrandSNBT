```python
# change format values 
df['NAMA PESERTA'] = df['NAMA PESERTA'].str.title() 
df['NAMA PESERTA']
```

```python
def getPTN(col_name): # function to split majoro (prodi) & university (univ/ptn)
    prodi = []
    ptn = []
    data = df[col_name]
    for d in data:
        x = d.rsplit(' (', 1)
        if len(x) == 1 :
            prodi.append('TIDAK MEMILIH')
            ptn.append('TIDAK MEMILIH')
        elif len(x) == 2:
            prodi.append(x[0])
            ptn.append(x[1][:-1])
        else:
            prodi.append(None)
            ptn.append(None)
    return prodi, ptn
```

```python
# splitting data
pil1, ptn1 = getPTN('Pilihan 1')
pil2, ptn2 = getPTN('Pilihan 2')
pil3, ptn3 = getPTN('Pilihan 3')
pil4, ptn4 = getPTN('Pilihan 4')
```

```python
# entered data

for p,c in zip([pil1, pil2, pil3, pil4],df.columns[2:6]):
    df[c] = p
    
df.head()
    
```

```python
df.rename(columns={'Pilihan 1':'Prodi 1',
          'Pilihan 2':'Prodi 2',
          'Pilihan 3':'Prodi 3',
          'Pilihan 4':'Prodi 4'}, inplace=True)
df.head()
```

```python
a = 1
for i,u in zip(range(3,11,2),[ptn1, ptn2, ptn3,ptn4]):
    df.insert(i, f'Univ {a}', u)
    a += 1
df.head()
```

### here we decided to make grade by score (SKOR TRYOUT UTBK)

###### < 450 grade C
###### 451 - 600 grade B
###### 601 - 750 grade A
###### dan > 750 grade Superior

```python
grade = pd.cut(df['SKOR TRYOUT UTBK'], 
               bins=[0,450,600,750,1000],
               labels=['C','B','A','Superior'])
```

```python
col_index = df.columns.to_list().index('SKOR TRYOUT UTBK')
df.insert(col_index + 1,'GRADE',grade)
```
